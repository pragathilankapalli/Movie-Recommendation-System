```python
import pandas as pd
import numpy as np
from content import Content
import json
import copy
```

```python
movies = pd.read_csv("ml-latest-small/linked_movies.csv")
```

```python
movie_genre = pd.DataFrame()
movie_language = pd.DataFrame()
```

```python
for i in range(0,len(movies)):
    movie =  movies.loc[i]
    movie_id = movie['movieId']
    genres = movie['genres'].split('|')
    languages = movie['languages'].split('|')
    df = pd.DataFrame({movie_id:np.ones(len(genres))}, index = genres)
    movie_genre = pd.concat([movie_genre, df], axis=1, sort=False)
    df = pd.DataFrame({movie_id:np.ones(len(languages))}, index = languages)
    movie_language = pd.concat([movie_language, df], axis=1, sort=False)
```

```python
ratings = pd.read_csv("ml-latest-small/ratings.csv")
```

```python
model = Content(ratings, movie_genre, movie_language)
```

```python
er = np.array(model.error)
rms_error = np.sqrt(np.mean(er**2))
```
